## 1.1 텐서 생성

### 파이토치의 텐서는 기본적인 파이썬의 iterable 객체를 활용하여 생성할 수 있다.

```python
import torch
import numpy as np
```

```python
# 리스트 -> 텐서
a = torch.tensor([0, 1, 2])

# 튜플 -> 텐서
b = ((1.0, 1.1), (1.2, 1.3))
b = torch.tensor(b)

# 넘파이 배열 -> 텐서
c = np.ones([2, 3])
c = torch.tensor(c)

print(f"Tensor a: {a}")
print(f"Tensor b: {b}")
print(f"Tensor c: {c}")
```

## 1.3 텐서 조작

### 텐서에서 index의 활용은 아래와 같다.

```python
x = torch.arange(0, 10)
print(x)
print(x[-1])
print(x[1:3]) 
print(x[:-2])
```

### 아래의 method를 통해 텐서를 reshape할 수 있다.

```python
# Reshaping
z = torch.arange(12).reshape(6, 2)
print(f"Original z (6, 2) : \n {z}")

# 2D -> 1D
z = z.flatten()
print(f"Flattened z: \n {z}")

# 1D -> 2D
z = z.reshape(3, 4)
print(f"Reshaped (3x4) z: \n {z}")
```

### 아래의 method를 통해 텐서들을 concatenate할 수 있다.

```python
x = torch.arange(12, dtype=torch.float32).reshape((3, 4))
y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])

# concatenate (row 기준)
cat_rows = torch.cat((x, y), dim=0)

# concatenate (columns 기준)
cat_cols = torch.cat((x, y), dim=1)

print('Concatenated by rows: shape{} \n {}'.format(list(cat_rows.shape), cat_rows))
print('\n Concatenated by colums: shape{}  \n {}'.format(list(cat_cols.shape), cat_cols))
```

### squeeze와 unsqueeze를 통해 singleton dimension을 없애거나 추가할 수 있다.

```python
x = torch.randn(1, 10)
print(x.shape)
print(f"x[0]: {x[0]}") 

# squeeze
x = x.squeeze(0)
print(x.shape)
print(f"x[0]: {x[0]}")

y = torch.randn(5, 5)
print(f"shape of y: {y.shape}")

# unsqueeze
y = y.unsqueeze(1) 
print(f"shape of y: {y.shape}")
```
